```python
import pylab as pl

import numpy as np

nstars = 20
data = np.loadtxt('data2.txt')
show = 10

time = data[:,0]
ke = data[:,1]
pe = data[:,2]
tot = data[:,3]

pos = []
for i in range(show):
    pos.append(data[:,(i*3+4):(i*3+7)])

pl.figure()


if __name__=='__main__':
    
    pl.subplot(221)
    pl.plot(time, ke, label='KE')
    pl.plot(time, pe, label='Potl. En.')
    pl.plot(time, tot, label='Total')
    pl.legend(frameon=False)

    pl.subplot(222)
    clrs = 'bgrcy'*show
    for i in range(show):
        #print(pos[i].shape
        pl.plot(pos[i][:,0],pos[i][:,1],c=clrs[i])
        pl.plot(pos[i][-1,0],pos[i][-1,1],clrs[i]+'o') # final pos
        pl.plot(pos[i][0,0],pos[i][0,1],clrs[i]+'^') # start pos

    pl.xlabel('x')
    pl.ylabel('y')

    pl.subplot(223)
    for i in range(show):
        #print pos[i].shape
        pl.plot(pos[i][:,0],pos[i][:,2])
        pl.plot(pos[i][-1,0],pos[i][-1,2],clrs[i]+'o') # final pos
        pl.plot(pos[i][0,0],pos[i][0,2],clrs[i]+'^') # start pos

    pl.xlabel('x')
    pl.ylabel('z')

    pl.show()
    #print pos[0][:,0]

```
