```python
#Getting energy data

def getData(string):
    string = string[:string.index('n')-2]
    string = ''.join(string)
    return int(string)

kinetic, potential, total = [], [], []

path = os.path.split(r'C:\Users\Syris Norelli\Documents\Processing Code\Functional_3D_N_Body\Python_Implementation\data.txt')
os.chdir(path[0])

f = open(path[1],'r')
iterator = True
print(f.readline())
while iterator:
    
    ke = list(f.readline()) #Kinetic Energy
    if 'n' in ke:
        kinetic.append(getData(ke))
    else:
        break
        
    pe = list(f.readline()) #Potential Energy
    if 'n' in ke:
        potential.append(getData(pe))
    else:
        break
        
    te = f.readline() #Total Energy
    if 'n' in te:
        total.append(getData(pe))
    else:
        break

print(kinetic[:100])
```
